# Load Dataset
Dataset dibaca menggunakan fungsi:
image_dataset_from_directory()

Dataset akan:
- di-resize menjadi 128x128
- dibagi menjadi training dan validation
- diproses sebagai citra RGB

```python
# =========================================================
# LOAD DATASET
# =========================================================

dataset_path = "dataset"

IMG_SIZE = 128
BATCH_SIZE = 8
AUTOTUNE = tf.data.AUTOTUNE

# Training Dataset
train_dataset = tf.keras.utils.image_dataset_from_directory(
    dataset_path,
    validation_split=0.2,
    subset="training",
    seed=42,
    image_size=(IMG_SIZE, IMG_SIZE),
    batch_size=BATCH_SIZE
)

# Validation Dataset
validation_dataset = tf.keras.utils.image_dataset_from_directory(
    dataset_path,
    validation_split=0.2,
    subset="validation",
    seed=42,
    image_size=(IMG_SIZE, IMG_SIZE),
    batch_size=BATCH_SIZE
)

# Nama class
CLASS_NAMES = train_dataset.class_names
NUM_CLASSES = len(CLASS_NAMES)

print("\nClass Names :", CLASS_NAMES)
print("Jumlah Kelas:", NUM_CLASSES)
```

# Preprocessing Data
Tahapan preprocessing yang dilakukan:

- Resize gambar menjadi 128x128
- Normalisasi piksel dari 0-255 menjadi 0-1
- Menyiapkan dataset untuk training CNN

```python
# =========================================================
# PREPROCESSING DATA
# =========================================================

def preprocess(image, label):

    # Resize gambar
    image = tf.image.resize(
        image,
        [IMG_SIZE, IMG_SIZE]
    )

    # Normalisasi piksel
    image = tf.cast(
        image,
        tf.float32
    ) / 255.0

    return image, label

# Terapkan preprocessing ke training
ds_train_proc = train_dataset.map(
    preprocess,
    num_parallel_calls=AUTOTUNE
)

# Terapkan preprocessing ke validation
ds_val_proc = validation_dataset.map(
    preprocess,
    num_parallel_calls=AUTOTUNE
)

print("=" * 50)
print("Preprocessing selesai")
print(f"Ukuran gambar : {IMG_SIZE}x{IMG_SIZE}")
print("Normalisasi   : 0-255 → 0-1")
print("=" * 50)
```

```python
# =========================================================
# CEK HASIL PREPROCESSING
# =========================================================

for image, label in ds_train_proc.take(1):

    print("Shape gambar :", image.shape)

    print("Tipe data    :", image.dtype)

    print("Nilai minimum:", image.numpy().min())

    print("Nilai maksimum:", image.numpy().max())
```
